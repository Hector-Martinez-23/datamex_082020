```python
resp5 = req.get('https://api.github.com/repos/ironhack-datalabs/scavenger/contents')
results5 = resp5.json()
```

```python
rutas=[]
for i in range(1,len(results5)):
    rutas.append(results5[i]['url'])
rutas
```

```python
results5[5]
```

```python
carpetas=[]
for carp in rutas:
    respcarp = req.get(carp)
    resultscarp = respcarp.json()
    for j in resultscarp:
        if 'scavengerhunt' in j['name']:
            carpetas.append([j['name'],j['download_url']])

carpetas
```

```python
carpetita=sorted(carpetas)
carpetita
```

```python
s=''
for k in carpetita:
    respfi = req.get(k[1])
    s=s+(respfi.text.replace('\n',' '))
```

```python
print(f'The message is: {s}')
```

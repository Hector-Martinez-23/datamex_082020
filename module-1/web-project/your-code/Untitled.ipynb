```python
cercanas=[]
def distancia(pa,es):
    for i in range(len(pa['Longitud'])):
        lon=float(pa['Longitud'][i])
        lat=float(pa['Latitud'][i])
        dists=[['a',1000],['b',1000],['c',1000]]
        for j in range(len(es)):    
            lon2=es['fields.longitud'][j]
            lat2=es['fields.latitud'][j]
            dis=(2*6372.795477598)*math.asin((math.sin((lat2-lat)/2))**2+math.cos(lat)*math.cos(lat2)*(math.sin((lon2-lon)/2))**2)
            d=[dists[0][1],dists[1][1],dists[2][1]]
            if dis<max(d):
                index=d.index(max(d))
                dists.pop(index)
                dists.append([es['recordid'][j],dis])
            else:
                pass
            dists.sort()
        cercanas.append(dists)
    return cercanas
```
